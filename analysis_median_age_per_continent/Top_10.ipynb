```python
# import dependencies
from path import Path
import pandas as pd
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from opencage.geocoder import OpenCageGeocode
%matplotlib inline
```

```python
# Import dataset & store into dataframe
merged_covid_df = pd.read_csv("../Resources/merged_covid_lon_lat.csv")
merged_covid_df
```

```python
# Set location as index
location_df = pd.DataFrame(merged_covid_df.set_index(["location"]))
location_df
```

```python
# Average new cases per location from highest to lowest
avg_newcases_df= location_df.groupby(by=["location"]).mean().sort_values(by="new_cases", ascending =False)
avg_newcases_df[0:5]
```

```python
newcases=avg_newcases_df.reset_index()
top5newcases =newcases[0:5]
top5newcases
```

```python
# Export top 5 new DataFrames to .csv
top5newcases.to_csv('../Resources/Top5NewCases.csv', index=False)
```

```python

# Average total deaths per location from highest to lowest
avg_totaldeaths=location_df.groupby(by=["location"]).mean().sort_values(by="total_deaths_updated",ascending=False)
avg_totaldeaths[0:5]
deaths=avg_totaldeaths.reset_index()[0:5]
deaths

```

```python
# Export top 5 total deaths DataFrames to .csv
deaths.to_csv('../Resources/Top5TotalDeaths.csv', index=False)
```

```python
# Average Stringency Index per location from highest to lowest
avg_StringencyIndex=location_df.groupby(by=["location"]).mean().sort_values(by="StringencyIndex_updated",ascending=False)
avg_StringencyIndex[0:5].reset_index()
```

```python
top5SI=avg_StringencyIndex[0:5].reset_index()
top5SI
```

```python
# Export top 5 Stringency index to .csv
top5SI.to_csv('../Resources/Top5StringencyIndex.csv', index=False)
```

```python
# Average Government Response per location from highest to lowest
Govt_Resp_avg=location_df.groupby(by=["location"]).mean().sort_values(by="GovernmentResponseIndex_updated",ascending=False)
GR=Govt_Resp_avg[0:5].reset_index()
```

```python
# Average Containment Index per location from Highest to lowest
ContainmentIndex_avg=location_df.groupby(by=["location"]).mean().sort_values(by="ContainmentHealthIndex_updated",ascending=False)
CI=ContainmentIndex_avg[0:5].reset_index()
```

```python
countries= merged_covid_df.location.unique()
countries
```

```python
country_lon_lat ={}
for country in countries:
    results = geocoder.geocode(country)
    country_lon_lat[country]={"lat":results[0]['geometry']['lat'], "lon": results[0]['geometry']['lng']}
country_lon_lat
```

```python
country_lon_lat['Aruba']['lat']
```

```python
lat = []
lon = []
for index, row in merged_covid_df.iterrows():
    location = row['location']
    for item in country_lon_lat:
        if location == item:
            lat.append(country_lon_lat[item]['lat'])
            lon.append(country_lon_lat[item]['lon'])
merged_covid_df['lat']=lat
merged_covid_df['lon']=lon
merged_covid_df.tail()
```
