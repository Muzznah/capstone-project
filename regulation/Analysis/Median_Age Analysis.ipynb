```python
# import dependencies
from path import Path
import pandas as pd
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# Import dataset & store into dataframe
merged_covid_df = pd.read_csv("../Resources/merged_covid_FINAL.csv")
merged_covid_df
```

```python
#Dataframe for Canada only

merged_covid_canada_df = pd.DataFrame(merged_covid_df.loc[merged_covid_df["location"]=="Canada"].reset_index())
merged_covid_canada_df = merged_covid_canada_df.drop(["index"],axis=1)
merged_covid_canada_df
```

```python
#  Dataframe filtered for Nigeria
merged_covid_Nigeria_df = pd.DataFrame(merged_covid_df.loc[merged_covid_df["location"]=="Nigeria"].reset_index())
merged_covid_Nigeria_df = merged_covid_Nigeria_df.drop(["index"],axis=1)
merged_covid_Nigeria_df
```

```python
#Set up axis-Canada
x_axis = merged_covid_canada_df["date_updated"]

y_axis1=merged_covid_canada_df["median_age"]
y_axis2 = merged_covid_canada_df["total_cases_updated"]

y_axis3 = merged_covid_canada_df["total_deaths_updated"]

y_axis4 = merged_covid_canada_df["new_cases"]
y_axis5 = merged_covid_canada_df["new_tests"]
y_axis6 = merged_covid_canada_df["new_deaths"]

```

```python
#Plot to compare  Median Age with New Deaths for Canada
fig, ax1 = plt.subplots()

color = 'tab:orange'
ax1.set_xlabel('Jan to Dec 2020')
ax1.set_ylabel('Median Age', color=color)
ax1.bar(x_axis, y_axis1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('New Deaths', color=color)  
ax2.plot(x_axis, y_axis6, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Median Age vs New Deaths-Canada")

fig.tight_layout()  
plt.show()
```

```python
# Set up Axis- Nigeria
x_axis0 = merged_covid_Nigeria_df["date_updated"]

y_axis01=merged_covid_Nigeria_df["median_age"]
y_axis02 = merged_covid_Nigeria_df["total_cases_updated"]

y_axis03= merged_covid_Nigeria_df["total_deaths_updated"]

y_axis04 = merged_covid_Nigeria_df["new_cases"]
y_axis05= merged_covid_Nigeria_df["new_tests"]
y_axis06 = merged_covid_Nigeria_df["new_deaths"]
```

```python
#Plot to compare Median Age with New Deaths- Nigeria
fig, ax1 = plt.subplots()

color = 'tab:orange'
ax1.set_xlabel('Jan to Dec 2020')
ax1.set_ylabel('Median Age', color=color)
ax1.bar(x_axis0, y_axis01, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('New Deaths', color=color)  
ax2.plot(x_axis0, y_axis06, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Median Age vs New Deaths-Nigeria")

fig.tight_layout()  
plt.show()
```
