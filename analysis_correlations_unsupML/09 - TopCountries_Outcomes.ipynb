# Get top & bottom 10 countries for (1) overall outcomes & for (2) time (days) from first case to max new cases
Correlate with specific stringency, health, econ indices

```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# load files
df = pd.read_csv("countryClusters.csv")
govResp = pd.read_csv('govRespOutcomes_withTests.csv')

display(df)
df.columns
```

```python
# make df for each outcome (class, country, outcome)
# max NC
maxNC_df = df[['Class','country','percent_maxNC']]
maxNC_df = maxNC_df.sort_values(by=['percent_maxNC'])
display(maxNC_df)
maxNC_df['rank'] = np.arange(1,163)
maxNC_df
```

```python
# make df for each outcome (class, country, outcome)
# max ND
maxND_df = df[['Class','country','percent_maxND']]
maxND_df = maxND_df.sort_values(by=['percent_maxND'])
display(maxND_df)

noZero_maxND = maxND_df.loc[maxND_df['percent_maxND'] != 0]
display(noZero_maxND)

zero_maxND = maxND_df.loc[maxND_df['percent_maxND'] == 0]
zero_maxND
```

```python
# make df for each outcome (class, country, outcome)
# NC3mnth
NC3mnth_df = df[['Class','country','percent_NC_3mnth']]
NC3mnth_df = NC3mnth_df.sort_values(by=['percent_NC_3mnth'])
display(NC3mnth_df)

noZero_NC3mnth = NC3mnth_df.loc[NC3mnth_df['percent_NC_3mnth'] != 0]
display(noZero_NC3mnth)

zero_NC3mnth = NC3mnth_df.loc[NC3mnth_df['percent_NC_3mnth'] == 0]
zero_NC3mnth
```

```python
# make df for each outcome (class, country, outcome)
# ND3mnth
ND3mnth_df = df[['Class','country','percent_ND_3mnth']]
ND3mnth_df = ND3mnth_df.sort_values(by=['percent_ND_3mnth'])
display(ND3mnth_df)

noZero_ND3mnth = ND3mnth_df.loc[ND3mnth_df['percent_ND_3mnth'] != 0]
display(noZero_ND3mnth)

zero_ND3mnth = ND3mnth_df.loc[ND3mnth_df['percent_ND_3mnth'] == 0]
zero_ND3mnth
```

```python
# make df for each outcome (class, country, outcome)
# TC3mnth
TC3mnth_df = df[['Class','country','percent_TC_3mnth']]
TC3mnth_df = TC3mnth_df.sort_values(by=['percent_TC_3mnth'])
TC3mnth_df['rank'] = np.arange(1,163)
display(TC3mnth_df)


noZero_TC3mnth = TC3mnth_df.loc[TC3mnth_df['percent_TC_3mnth'] != 0]
display(noZero_TC3mnth)

zero_TC3mnth = TC3mnth_df.loc[TC3mnth_df['percent_TC_3mnth'] == 0]
zero_TC3mnth
```

```python
# make df for each outcome (class, country, outcome)
# TD3mnth
TD3mnth_df = df[['Class','country','percent_TD_3mnth']]
TD3mnth_df = TD3mnth_df.sort_values(by=['percent_TD_3mnth'])
display(TD3mnth_df)

noZero_TD3mnth = TD3mnth_df.loc[TD3mnth_df['percent_TD_3mnth'] != 0]
display(noZero_TD3mnth)

zero_TD3mnth = TD3mnth_df.loc[TD3mnth_df['percent_TD_3mnth'] == 0]
zero_TD3mnth
```

```python
# make df for each outcome (class, country, outcome)
# firstNCToMaxNC
firstNCToMaxNC_df = df[['Class','country','interval_firstNCToMaxNC']]
firstNCToMaxNC_df = firstNCToMaxNC_df.sort_values(by=['interval_firstNCToMaxNC'], ascending = False)
firstNCToMaxNC_df['rank'] = np.arange(1,163)
display(firstNCToMaxNC_df)

noZero_firstNCToMaxNC = firstNCToMaxNC_df.loc[firstNCToMaxNC_df['interval_firstNCToMaxNC'] != 0]
display(noZero_firstNCToMaxNC)

zero_firstNCToMaxNC = firstNCToMaxNC_df.loc[firstNCToMaxNC_df['interval_firstNCToMaxNC'] == 0]
zero_firstNCToMaxNC
```

```python
# make df for each outcome (class, country, outcome)
# firstNCToMaxND
firstNCToMaxND_df = df[['Class','country','interval_firstNCToMaxND']]
firstNCToMaxND_df = firstNCToMaxND_df.sort_values(by=['interval_firstNCToMaxND'], ascending = False)
display(firstNCToMaxND_df)

noZero_firstNCToMaxND_df = firstNCToMaxND_df.loc[firstNCToMaxND_df['interval_firstNCToMaxND'] != 0]
display(noZero_firstNCToMaxND_df)

zero_firstNCToMaxND = firstNCToMaxND_df.loc[firstNCToMaxND_df['interval_firstNCToMaxND'] == 0]
zero_firstNCToMaxND
```

```python
# preview dfs
maxNC_df.set_index('country', inplace=True)
display(maxNC_df)

TC3mnth_df.set_index('country', inplace=True)
display(TC3mnth_df)

firstNCToMaxNC_df.set_index('country', inplace=True)
display(firstNCToMaxNC_df)
```

# Top10 & Bottom10 Countries for All Outcomes 
### (Outcomes: percent_maxNC, percent_TC_3mnth, interval_firstNCToMaxNC)

```python
# join dfs from above
join_df = maxNC_df.join(TC3mnth_df, how="inner", lsuffix='_maxNC', rsuffix='_TC3mnth')
join_df

join3_df = join_df.join(firstNCToMaxNC_df, how="inner")
join3_df
join3_df['rankSum'] = join3_df['rank_maxNC'] + join3_df['rank_TC3mnth'] + join3_df['rank']
join3_df = join3_df.sort_values(by='rankSum')
top10_allOutcomes = join3_df.head(10)
bottom10_allOutcomes = join3_df.tail(10)

print("Top 10 All Outcomes")
display(top10_allOutcomes)
print("")
print("Bottom 10 All Outcomes")
display(bottom10_allOutcomes)
```

```python
# ranking for interval first case to max NC
top10firstNCToMaxNC = firstNCToMaxNC_df.head(10)
print("Top 10 Interval First Case to Max New Cases")
display(top10firstNCToMaxNC)
print("")

print("Bottom 10 Interval First Case to Max New Cases")
bottom10firstNCToMaxNC = firstNCToMaxNC_df.tail(10)
display(bottom10firstNCToMaxNC)
```

```python
# top 10 all outcomes
# drop columns so left with rankSum (sum of all rankngs for all outcomes)

display(top10_allOutcomes)
top10_allOutcomes.columns
top10_allOutcomes.drop(columns=['Class_maxNC', 'percent_maxNC', 'rank_maxNC', 'Class_TC3mnth',
       'percent_TC_3mnth', 'rank_TC3mnth','interval_firstNCToMaxNC','rank'], inplace=True)

top10_allOutcomes = top10_allOutcomes.reset_index()
top10_allOutcomes
```

```python
# bottom10_allOutcomes
# drop columns so left with rankSum (sum of all rankngs for all outcomes)

display(bottom10_allOutcomes)
bottom10_allOutcomes.columns
bottom10_allOutcomes.drop(columns=['Class_maxNC', 'percent_maxNC', 'rank_maxNC', 'Class_TC3mnth',
       'percent_TC_3mnth', 'rank_TC3mnth','interval_firstNCToMaxNC','rank'], inplace=True)
bottom10_allOutcomes

bottom10_allOutcomes = bottom10_allOutcomes.reset_index()
bottom10_allOutcomes
```

```python
# preview df
display(govResp)
govResp.columns
```

```python
# join top 10 rankSum with specific gov measures in govResp df
top10All_merge = top10_allOutcomes.merge(govResp, on="country", how="inner")
top10All_merge.columns

top10All_merge.drop(columns=['primary_key', 'iso_code', 'population',
       'population_density', 'median_age', 'gdp_per_capita', 'date_minGR',
       'date_maxGR', 'date_firstNC', 'date_maxNC', 'date_maxND',
       'date_3mnthAfter', 'interval_GRRamp', 'interval_GRToFirstNC',
       'interval_firstNCToMaxNC', 'interval_firstNCToMaxND',
       'interval_maxNCToMaxND', 'index_MaxGR', 'index_avgMaxGRStartApr1',
       'change_GRRamp', 'index_string', 'index_health', 'index_econ',
       'value_maxNC', 'value_maxND', 'value_NC_3mnth', 'value_TC_3mnth',
       'value_ND_3mnth', 'value_TD_3mnth', 'percent_maxNC', 'percent_maxND',
       'percent_NC_3mnth', 'percent_ND_3mnth', 'percent_TC_3mnth',
       'percent_TD_3mnth', 'percent_newTestsMean', 'value_totTestsSlope',
       'mobil_residential', 'mobil_outdoors', 'mobil_retailRecreation',
       'mobil_parks', 'mobil_GroceryPharmacy', 'mobil_transitStations',
       'mobil_workplaces'], inplace=True)
display(top10All_merge)
top10All_merge['top10_allOutcomes'] = "top10_allOutcomes"
top10All_merge

top10_grp = top10All_merge.groupby(by="top10_allOutcomes").mean()
top10_grp
```

```python
# join bottom 10 rankSum with specific gov measures in govResp df
bottom10All_merge = bottom10_allOutcomes.merge(govResp, on="country", how="inner")
bottom10All_merge.columns

bottom10All_merge.drop(columns=['primary_key', 'iso_code', 'population',
       'population_density', 'median_age', 'gdp_per_capita', 'date_minGR',
       'date_maxGR', 'date_firstNC', 'date_maxNC', 'date_maxND',
       'date_3mnthAfter', 'interval_GRRamp', 'interval_GRToFirstNC',
       'interval_firstNCToMaxNC', 'interval_firstNCToMaxND',
       'interval_maxNCToMaxND', 'index_MaxGR', 'index_avgMaxGRStartApr1',
       'change_GRRamp', 'index_string', 'index_health', 'index_econ',
       'value_maxNC', 'value_maxND', 'value_NC_3mnth', 'value_TC_3mnth',
       'value_ND_3mnth', 'value_TD_3mnth', 'percent_maxNC', 'percent_maxND',
       'percent_NC_3mnth', 'percent_ND_3mnth', 'percent_TC_3mnth',
       'percent_TD_3mnth', 'percent_newTestsMean', 'value_totTestsSlope',
       'mobil_residential', 'mobil_outdoors', 'mobil_retailRecreation',
       'mobil_parks', 'mobil_GroceryPharmacy', 'mobil_transitStations',
       'mobil_workplaces'], inplace=True)
display(bottom10All_merge)
bottom10All_merge['bottom10_allOutcomes'] = "bottom10_allOutcomes"
bottom10All_merge

bottom10_grp = bottom10All_merge.groupby(by="bottom10_allOutcomes").mean()
bottom10_grp
```

```python
# make a df that contains "all" countries with specific government measures
join3_dfCopy = join3_df.copy()
join3_dfCopy = join3_dfCopy.reset_index()
join3_dfCopy

allCountries_merge = join3_dfCopy.merge(govResp, on="country", how="inner")
allCountries_merge.columns

allCountries_merge.drop(columns=['country', 'Class_maxNC', 'percent_maxNC_x', 'rank_maxNC',
       'Class_TC3mnth', 'percent_TC_3mnth_x', 'rank_TC3mnth',
       'interval_firstNCToMaxNC_x', 'rank', 'primary_key',
       'iso_code', 'population', 'population_density', 'median_age',
       'gdp_per_capita', 'date_minGR', 'date_maxGR', 'date_firstNC',
       'date_maxNC', 'date_maxND', 'date_3mnthAfter', 'interval_GRRamp',
       'interval_GRToFirstNC', 'interval_firstNCToMaxNC_y',
       'interval_firstNCToMaxND', 'interval_maxNCToMaxND', 'index_MaxGR',
       'index_avgMaxGRStartApr1', 'change_GRRamp', 'index_string',
       'index_health', 'index_econ', 'value_maxNC', 'value_maxND',
       'value_NC_3mnth', 'value_TC_3mnth', 'value_ND_3mnth', 'value_TD_3mnth',
       'percent_maxNC_y', 'percent_maxND', 'percent_NC_3mnth',
       'percent_ND_3mnth', 'percent_TC_3mnth_y', 'percent_TD_3mnth',
       'percent_newTestsMean', 'value_totTestsSlope', 'mobil_residential',
       'mobil_outdoors', 'mobil_retailRecreation', 'mobil_parks',
       'mobil_GroceryPharmacy', 'mobil_transitStations', 'mobil_workplaces'], inplace=True)
display(allCountries_merge)
allCountries_merge['allCountries_allOutcomes'] = "allCountries_allOutcomes"
allCountries_merge

allCountries_grp = allCountries_merge.groupby(by="allCountries_allOutcomes").mean()
allCountries_grp
```

```python
# concatenate top 10, bottom 10, and all countries df to show ranksum and specific government measures
allOutcomes = pd.concat([top10_grp,bottom10_grp, allCountries_grp])
allOutcomes
```

```python
# drop unneeded "Class" column
allOutcomes.drop(columns=["Class"], inplace=True)
display(allOutcomes)
allOutcomes.columns
```

```python
# convert to integers
allOutcomes_float = allOutcomes[['rankSum', 'C1_School closing', 'C1_Flag','C2_Workplace closing','C2_Flag',
       'C3_Cancel public events','C3_Flag', 'C4_Restrictions on gatherings','C4_Flag',
       'C5_Close public transport', 'C5_Flag','C6_Stay at home requirements','C6_Flag',
       'C7_Restrictions on internal movement','C7_Flag',
       'C8_International travel controls', 'E1_Income support','E1_Flag',
       'E2_Debt/contract relief', 'H1_Public information campaigns','H1_Flag',
       'H2_Testing policy', 'H3_Contact tracing']]
allOutcomes_int = allOutcomes_float.astype(int)
allOutcomes_int
```
