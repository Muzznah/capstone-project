# This notebook created T_canUS.csv

```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
# load files
gov_reg_data = "government_regulation_FINAL.csv"
covid_stats_data = "covid_stats_FINAL.csv"
merged_covid_data = "merged_covid_FINAL.csv"

gov_reg = pd.read_csv(gov_reg_data)
covid_stats = pd.read_csv(covid_stats_data)
merged_covid = pd.read_csv(merged_covid_data)
```

# Canada & US

```python
# canada df
canada_df = merged_covid.loc[merged_covid['location'] == 'Canada']
canada_df.set_index(['date_updated'], inplace=True)
display(canada_df)

canada_df.to_csv("canada.csv")
# date range from Dec 31, 2019 to Aug 31, 2020

# us df
us_df = merged_covid.loc[merged_covid['location'] == 'United States']
us_df.set_index(['date_updated'], inplace=True)
us_df

us_df.to_csv("us.csv")
# date range is same as Canada 
```

```python
# preview dfs
display(canada_df.describe())
display(us_df.describe())
```

```python
# join dfs
canUS_df = canada_df.join(us_df, how="inner", lsuffix="_CAN", rsuffix="_US")
canUS_df = canUS_df.reset_index()

display(canUS_df)
canUS_df.columns
```

# Scale Canada

```python
# scale: new cases, new deaths, new tests, total cases updated, total deaths updated, total tests updated
# for both Canada & US

scaler = MinMaxScaler()

# fit scaler to df
colsNeedScale_df = canUS_df[['new_cases_CAN', 'new_deaths_CAN', 'new_tests_CAN',
       'total_cases_updated_CAN', 'total_deaths_updated_CAN',
       'total_tests_updated_CAN','new_cases_US', 'new_deaths_US', 'new_tests_US',
       'total_cases_updated_US', 'total_deaths_updated_US',
       'total_tests_updated_US']]

scaler.fit(colsNeedScale_df)
print(scaler.data_max_) # highest max value is total_tests_updated_US

# create series for each column that needs scaling
new_cases_CAN = canUS_df['new_cases_CAN']
new_deaths_CAN = canUS_df['new_deaths_CAN']
new_tests_CAN = canUS_df['new_tests_CAN']
total_cases_updated_CAN = canUS_df['total_cases_updated_CAN']
total_deaths_updated_CAN = canUS_df['total_deaths_updated_CAN']
total_tests_updated_CAN = canUS_df['total_tests_updated_CAN']

new_cases_US = canUS_df['new_cases_US']
new_deaths_US = canUS_df['new_deaths_US']
new_tests_US = canUS_df['new_tests_US']
total_cases_updated_US = canUS_df['total_cases_updated_US']
total_deaths_updated_US = canUS_df['total_deaths_updated_US']
total_tests_updated_US = canUS_df['total_tests_updated_US']
```

```python
# prep total_tests_updated_CAN to be fit by scaler
total_tests_updated_CAN = np.asarray(total_tests_updated_CAN)
# display(total_tests_updated_CAN)

total_tests_updated_CAN = total_tests_updated_CAN.reshape(-1,1)
# display(total_tests_updated_CAN)

# re-scale based on column with max value (total_tests_updated_CAN)
scaler.fit(total_tests_updated_CAN)
print(scaler.data_max_)

# transform total_tests_updated_CAN
T_total_tests_updated_CAN = scaler.transform(total_tests_updated_CAN)
print(T_total_tests_updated_CAN)
```

```python
# function that: apply asarray, reshape, transform to other columns
def scale(col):
    col = np.asarray(col)
    col = col.reshape(-1,1)
    col = scaler.transform(col)
    return col

# apply function
T_new_cases_CAN = scale(new_cases_CAN)
T_new_deaths_CAN = scale(new_deaths_CAN)
T_new_tests_CAN = scale(new_tests_CAN)
T_total_cases_updated_CAN = scale(total_cases_updated_CAN)
T_total_deaths_updated_CAN = scale(total_deaths_updated_CAN)

# drop old columns in T_canUS_df
T_canUS_df = canUS_df.copy()
T_canUS_df.drop(columns=['new_cases_CAN', 'new_deaths_CAN', 'new_tests_CAN',
       'total_cases_updated_CAN', 'total_deaths_updated_CAN',
       'total_tests_updated_CAN','new_cases_US', 'new_deaths_US', 'new_tests_US',
       'total_cases_updated_US', 'total_deaths_updated_US',
       'total_tests_updated_US'], inplace=True)

# attach new columns
T_canUS_df['T_new_cases_CAN'] = T_new_cases_CAN
T_canUS_df['T_new_deaths_CAN'] = T_new_deaths_CAN
T_canUS_df['T_new_tests_CAN'] = T_new_tests_CAN
T_canUS_df['T_total_cases_updated_CAN'] = T_total_cases_updated_CAN
T_canUS_df['T_total_deaths_updated_CAN'] = T_total_deaths_updated_CAN
T_canUS_df['T_total_tests_updated_CAN'] = T_total_tests_updated_CAN
```

# Scale the US

```python
# prep total_tests_updated_US to be fit by scaler
total_tests_updated_US = np.asarray(total_tests_updated_US)
# display(total_tests_updated_US)

total_tests_updated_US = total_tests_updated_US.reshape(-1,1)
# display(total_tests_updated_US)

# re-scale based on column with max value (total_tests_updated_US)
scaler.fit(total_tests_updated_US)
print(scaler.data_max_)

# transform total_tests_updated_US
T_total_tests_updated_US = scaler.transform(total_tests_updated_US)
print(T_total_tests_updated_US)
```

```python
T_new_cases_US = scale(new_cases_US)
T_new_deaths_US = scale(new_deaths_US)
T_new_tests_US = scale(new_tests_US)
T_total_cases_updated_US = scale(total_cases_updated_US)
T_total_deaths_updated_US = scale(total_deaths_updated_US)
T_total_tests_updated_US = scale(total_tests_updated_US)
```

```python
T_canUS_df['T_new_cases_US'] = T_new_cases_US
T_canUS_df['T_new_deaths_US'] = T_new_deaths_US
T_canUS_df['T_new_tests_US'] = T_new_tests_US
T_canUS_df['T_total_cases_updated_US'] = T_total_cases_updated_US
T_canUS_df['T_total_deaths_updated_US'] = T_total_deaths_updated_US
T_canUS_df['T_total_tests_updated_US'] = T_total_tests_updated_US
```
