# Modification of dataset to include all feature & target variables for analysis

```python
# import dependencies
import pandas as pd
import numpy as np
from statistics import mean
from datetime import timedelta
```

```python
# load files
govReg = pd.read_csv("OxCGRT_latest(Aug31,2020).csv")
gdp = pd.read_csv("owid-covid-data(Aug31,2020).csv")
mobil = pd.read_csv("regional_mobility.csv")

merged_covid = pd.read_csv("merged_covid_FINAL.csv")

string = pd.read_csv("stringency.csv")
health = pd.read_csv("health.csv")
econ = pd.read_csv("econ_support.csv")
avgMaxGovResp = pd.read_csv("gov_resp.csv")
```

```python
# preview data
display(merged_covid)
display(avgMaxGovResp)
display(list(gdp.columns))

# find out number of countries in merged_covid: 164 countries
len(list(merged_covid['location'].unique()))
```

```python
# sort by location, then date
merged_covid = merged_covid.sort_values(by=['location','date_updated'])
display(merged_covid)
display(merged_covid.describe())

# reset index
merged_covid = merged_covid.reset_index()
merged_covid.drop(columns=['index'], inplace=True)
merged_covid
```

## Find avg. minimum government response for each country

```python
# for loop to find avg. min. gov response
# avg. min. gov response is 14-15

countries = list(merged_covid['location'].unique())
# length of "countries" list
print(len(countries))

min_list = []

for country in countries:
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    govResp = countryLoc['GovernmentResponseIndex_updated']
    govResp_min = govResp.min(axis=0)
    min_list.append(govResp_min)

# view min. gov response of all countries
print(min_list)
# ensure obtained min. gov response for all countries (there should be 164)
print(len(min_list))

# avg. min. gov response
mean = mean(min_list)
mean
```

## Get date of max. value of gov response for each country

```python
# for loop to get index of max. value of gov response
maxIndex_list = []

for country in countries:
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    govResp = countryLoc['GovernmentResponseIndex_updated']
    govResp_maxIndex = govResp.idxmax()
    maxIndex_list.append(govResp_maxIndex)
    
print(maxIndex_list)

# ensure got index for all countries
print(len(maxIndex_list))
```

```python
# for loop to get max. value of gov response
maxValue_list = []

for country in countries:
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    govResp = countryLoc['GovernmentResponseIndex_updated']
    govResp_maxValue = govResp.max()
    maxValue_list.append(govResp_maxValue)
    
print(maxValue_list)

# ensure got max. value for all countries
print(len(maxValue_list))
```

```python
# for loop to get date of first max. gov. response
dateMax_list = []

for index in maxIndex_list:
    dateMax = merged_covid.iloc[index,1]
    dateMax_list.append(dateMax)

dateMax_list
```

## Get date of first min. gov response that is >14 

```python
# for loop
dateMin_list = []

for country in countries:
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    minLoc = countryLoc.loc[countryLoc['GovernmentResponseIndex_updated'] > 14]
    
    indexMin = minLoc.index[0]
    dateMin = merged_covid.iloc[indexMin,1]
    dateMin_list.append(dateMin)

print(dateMin_list)

# ensure got date for each country
print(len(dateMin_list))
```

## Get  (1) time (days) from min. gov response (>14) to max. gov response
## and (2) difference between max & min government response 
## for each country 

```python
grp_df = merged_covid.groupby('location').max()
grp_df['minDate'] = dateMin_list
grp_df['maxDate'] = dateMax_list
grp_df['maxValue'] = maxValue_list

# get difference between max & min government response for each country
grp_df['changeGovResp'] = grp_df['maxValue'] - 14

display(grp_df)
# check data types of columns
grp_df.dtypes
```

```python
# convert dates to datetime
grp_df['minDate'] = pd.to_datetime(grp_df['minDate'])
grp_df['maxDate'] = pd.to_datetime(grp_df['maxDate'])

# get time (days) from min. to max. gov response
grp_df['ramp_up_duration'] = grp_df['maxDate'] - grp_df['minDate']
display(grp_df.dtypes)

# ensure data types have been converted
grp_df
```

```python
# reset index
grp_df = grp_df.reset_index()
grp_df

# reorder dataframe
rampUp_df = grp_df[['location','primary_key_reaction','iso_code','minDate','maxDate','maxValue','changeGovResp','ramp_up_duration','population','population_density','median_age']]
rampUp_df
```

## Get date of first covid case for each country

```python
# for loop to get index of first covid case for each country
index_list = []

for country in countries:
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    nonZero_df = countryLoc[(countryLoc['new_cases'] != 0)]
    if nonZero_df.empty == False:        
        index = nonZero_df.index[nonZero_df['new_cases'] > 0].to_list()[0]
    index_list.append(index)

index_list
```

```python
# using the index, get the date for the first covid case for each country
newCases_startDate = []

for index in index_list:
    startDate = merged_covid.iloc[index,1]
    newCases_startDate.append(startDate)

newCases_startDate
```

```python
# add date of first case for each country to the dataframe
rampUp_df['newCases_startDate'] = newCases_startDate
rampUp_df['newCases_startDate'] = pd.to_datetime(rampUp_df['newCases_startDate'])

# check the data types
rampUp_df.dtypes
```

## Determine the time (days) from start of gov. response to first covid case for each country
### Positive number means gov resp started AFTER first case
### Negative number means gov resp started BEFORE first case

```python
rampUp_df['CaseVsResp_Interval'] = rampUp_df['newCases_startDate'] - rampUp_df['minDate']

# check dataframe
display(rampUp_df)
display(rampUp_df.describe())
list(rampUp_df.columns)
```

## include avg. max. gov response starting Apr 1, 2020 (and continuing until Aug 31, 2020, inclusive)

```python
# preview dataframe (df)
display(avgMaxGovResp)

join_df = rampUp_df.merge(avgMaxGovResp, how="inner", on="location")
join_df = join_df.rename(columns={"GovernmentResponseIndex_updated":"AvgMaxGovResp_StartApr1"})

# ensure join & renaming occurred properly
join_df
```

## Find date & value of max. new cases (peak of bell curve) for each country

```python
# for loop
indexMaxNewCases_list = []
valueMaxNewCases_list = []
dateMaxNewCases_list = []

for country in countries:
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    maxNewCases_index = countryLoc['new_cases'].idxmax()
    maxNewCases_value = countryLoc['new_cases'].max()
    
    indexMaxNewCases_list.append(maxNewCases_index)
    valueMaxNewCases_list.append(maxNewCases_value)

# print to preview index and value of max. new cases for each country
# print(indexMaxNewCases_list)
# print(valueMaxNewCases_list)

for index in indexMaxNewCases_list:
    maxNewCases_date = merged_covid.iloc[index,1]
    dateMaxNewCases_list.append(maxNewCases_date)

# list of dates for max. value of new cases for each country
dateMaxNewCases_list
```

## Find time (days) from first case to max. new cases for each country

```python
join_df['MaxNewCases_date'] = dateMaxNewCases_list
join_df['MaxNewCases_value'] = valueMaxNewCases_list

join_df['MaxNewCases_date'] = pd.to_datetime(join_df['MaxNewCases_date'])
join_df['StartToMaxNewCases_Interval'] = join_df['MaxNewCases_date'] - join_df['minDate']

# ensure proper join and check data types (dtypes)
display(join_df)
join_df.dtypes
```

## Find value and date of max new deaths for each country

```python
# for loop
indexMaxNewDeaths_list = []
valueMaxNewDeaths_list = []
dateMaxNewDeaths_list = []

for country in countries:
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    maxNewDeaths_index = countryLoc['new_deaths'].idxmax()
    maxNewDeaths_value = countryLoc['new_deaths'].max()
    
    indexMaxNewDeaths_list.append(maxNewDeaths_index)
    valueMaxNewDeaths_list.append(maxNewDeaths_value)

# print to preview index and value of max. new cases for each country
# print(indexMaxNewCases_list)
# print(valueMaxNewCases_list)

for index in indexMaxNewDeaths_list:
    maxNewDeaths_date = merged_covid.iloc[index,1]
    dateMaxNewDeaths_list.append(maxNewDeaths_date)

# list of dates of max. value of new deaths for each country
dateMaxNewDeaths_list
```

## Time (days) from max. new cases to max. new deaths for each country

```python
join_df['MaxNewDeaths_date'] = dateMaxNewDeaths_list
join_df['MaxNewDeaths_value'] = valueMaxNewDeaths_list

join_df['MaxNewDeaths_date'] = pd.to_datetime(join_df['MaxNewDeaths_date'])

join_df['MaxNewCasesToMaxNewDeaths_Interval'] = join_df['MaxNewDeaths_date'] - join_df['MaxNewCases_date']

# check df & dtypes
display(join_df)
join_df.dtypes
```

## Time (days) from first new case to max. new deaths for each country

```python
join_df['FirstNewCaseToMaxNewDeaths'] = join_df['MaxNewDeaths_date'] - join_df['newCases_startDate']

# check df after addition of the new column
display(join_df)
```

## Find the date 3 months from the date of the first case for each country
assume 1 month = 30 days

```python
join_df['3mth_date'] = join_df['newCases_startDate'] + timedelta(days = 30)

# check df and dtypes
display(join_df)
join_df.dtypes
```

## Find max. new cases, & total cases and total deaths 3 months (MO) after first case for each country

```python
# series of dates 3 MO after the first case for each country
date3mnth = join_df['3mth_date']
date3mnth
```

```python
# convert "date_updated" in merged_covid to datetime
merged_covid['date_updated'] = pd.to_datetime(merged_covid['date_updated'])
merged_covid.dtypes
```

```python
# for loop
countries3mnth_array = []

for i in range(0,164):
    country = countries[i]
    date3mnth_country = date3mnth[i]
    country3mnth_dict = {}
    
    countryLoc = merged_covid.loc[merged_covid['location'] == country]
    maxNewCases_3mnth_df = countryLoc[(countryLoc['date_updated'] >= date3mnth_country)]
    
    if maxNewCases_3mnth_df.empty == False:
        newCases = maxNewCases_3mnth_df.iloc[0,8]
        totCases = maxNewCases_3mnth_df.iloc[0,14]
        newDeaths = maxNewCases_3mnth_df.iloc[0,10] 
        totDeaths = maxNewCases_3mnth_df.iloc[0,16] 
        
        country3mnth_dict['country'] = country 
        country3mnth_dict['newCases'] = newCases
        country3mnth_dict['totCases'] = totCases
        country3mnth_dict['newDeaths'] = newDeaths
        country3mnth_dict['totDeaths'] = totDeaths

        countries3mnth_array.append(country3mnth_dict)

countries3mnth_array
```

```python
# create 3 mnths later df:
# index = length of array
# columns = country, newCases, totCases, newDeaths, totDeaths

# check there are 164 values corresponding to 164 countries
print(len(countries3mnth_array))

countries = []
newCases_list = []
totCases_list = []
newDeaths_list = []
totDeaths_list = []

for item in countries3mnth_array:
    country = item['country']
    countries.append(country)
    
    newCases = item['newCases']
    newCases_list.append(newCases)
    
    totCases = item['totCases']
    totCases_list.append(totCases)
    
    newDeaths = item['newDeaths']
    newDeaths_list.append(newDeaths)
    
    totDeaths = item['totDeaths']
    totDeaths_list.append(totDeaths)
    
# print to preview each list
# print(countries)
# print(newCases_list)
# print(totCases_list)
# print(newDeaths_list)
# print(totDeaths_list)

df_dict = {"location": countries,
           "newCases_3mnth": newCases_list,
           "totCases_3mnth": totCases_list,
           "newDeaths_3mnth": newDeaths_list,
           "totDeaths_3mnth": totDeaths_list
          }

indexDF = np.array(range(0,164))

# make df
df_3mnth = pd.DataFrame.from_dict(df_dict)

# preview created df
df_3mnth
```

## Join df with 3 mnth data to previous df

## Include mobility in the previous df
Will use avg. mobility for each country over a 3-MO duration (from min. gov resp to 3 mnths after that)

```python
# mobility (avg. mobility outdoors and indoors over the 3 MO)
# sort by country, then date
mobil.sort_values(by=['country','date'])
mobil.drop(columns="Unnamed: 0", inplace=True)
display(mobil)
```

```python
# discard countries in mobility df that aren't present in the previous df
mobilCountry_list = list(mobil['country'].unique())
mobilCountry_list

for country in mobilCountry_list:
    if country not in countries:
        lessCountries_df = mobil[(mobil['country'] != country)]

display(lessCountries_df)
len(list(lessCountries_df['country'].unique()))
```

```python
# drop rows if date is BEFORE date of min. gov response

minDate_df = mergeEcon[['location','minDate']]
minDate_df.rename(columns={'location':'country'}, inplace=True)
minDate_df

# determine date 3 mnths AFTER date of min. gov response
minDate_df['3mnth'] = minDate_df['minDate'] + timedelta(days = 90)
minDate_df
```

```python
# merge "3-MO-later" date with rest of mobility df
mobilDate_df = lessCountries_df.merge(minDate_df, on='country', how='left')
mobilDate_df['date'] = pd.to_datetime(mobilDate_df['date'])

# ensure merged correctly with proper dtypes
display(mobilDate_df)
mobilDate_df.dtypes
```

```python
# remove dates AFTER the 3 MO date

newMobilCountry_list = list(mobilDate_df['country'].unique())
len(newMobilCountry_list)

countryLoc_list = []

for i in range(0,134):
    country = newMobilCountry_list[i]
    countryLoc = mobilDate_df.loc[mobilDate_df['country'] == country]
    datesAfterGR_df = countryLoc[(countryLoc['date'] > countryLoc['minDate'])]
    dates3mnthAfterMinDate_df = datesAfterGR_df[datesAfterGR_df['date'] <= datesAfterGR_df['3mnth']]
    
    countryLoc_list.append(dates3mnthAfterMinDate_df)
    mobil3mnth_df = pd.concat(countryLoc_list)

# check the df
display(mobil3mnth_df)

# save df
mobil3mnth_df.to_csv("mobil_3mnth.csv", index=False)
```

```python
# get avg. change in mobility outdoors and indoors 
avgMobil_df = mobil3mnth_df.groupby(['country']).mean()

# check df
display(avgMobil_df)
avgMobil_df.dtypes
```

```python
# get the avg. of all "outdoors" mobility columns
avgMobil_df['avgOutdoorMobil'] = (avgMobil_df['retail_and_recreation'] + avgMobil_df['grocery_and_pharmacy'] + avgMobil_df['parks'] + avgMobil_df['transit_stations'] + avgMobil_df['workplaces'])/5

avgMobil_df = avgMobil_df.reset_index()
avgMobil_df.rename(columns={"country":"location"}, inplace=True)
avgMobil_df
```

```python
# merge mobility df with previous df
merge = mergeEcon.merge(avgMobil_df, on='location', how='left')

# check merged df
display(merge)
merge.columns.to_list()
```
