# Take log of target variables for RStudio analysis

```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# load files
GRO = pd.read_csv("countryClusters.csv")

# preview df
pd.set_option("display.max_columns", 73)
display(GRO)
GRO.columns
```

```python
# apply log function
GRO['logPercent_maxNC'] = np.log10(GRO['percent_maxNC'])
GRO['logPercent_maxND'] = np.log10(GRO['percent_maxND'])
GRO['logPercent_NC_3mnth'] = np.log10(GRO['percent_NC_3mnth'])
GRO['logPercent_ND_3mnth'] = np.log10(GRO['percent_ND_3mnth'])
GRO['logPercent_TC_3mnth'] = np.log10(GRO['percent_TC_3mnth'])
GRO['logPercent_TD_3mnth'] = np.log10(GRO['percent_TD_3mnth'])
GRO['logInterval_firstNCToMaxNC'] = np.log10(GRO['interval_firstNCToMaxNC'])
GRO['logInterval_firstNCToMaxND'] = np.log10(GRO['interval_firstNCToMaxND'])

display(GRO)

# see number of null rows
GRO.isnull().sum()
```

```python
# # Replacing infinite with nan 
GRO.replace([np.inf, -np.inf], np.nan, inplace=True)
display(GRO)

# view columns
GRO.columns
```

```python
# create new df that drops NaN
# include PC1, PC2, PC3, Class, country, and the target variables
target_df = GRO[['country','PC 1', 'PC 2', 'PC 3', 'Class', 'logPercent_maxNC',
       'logPercent_maxND', 'logPercent_NC_3mnth', 'logPercent_ND_3mnth',
       'logPercent_TC_3mnth', 'logPercent_TD_3mnth','logInterval_firstNCToMaxNC','logInterval_firstNCToMaxND']]
display(target_df)

# view null rows
target_df.isnull().sum()
```

```python
# describe df
target_df.describe()
```

```python
# create dfs of each target column with null values
log_maxND_df = target_df[['country','PC 1', 'PC 2', 'PC 3', 'Class',
       'logPercent_maxND']]
log_maxND_df = log_maxND_df.dropna()
display(log_maxND_df)

# save to csv
log_maxND_df.to_csv("log_maxND.csv")
```

```python
# create dfs of each target column with null values
log_NC_3mnth_df = target_df[['country','PC 1', 'PC 2', 'PC 3', 'Class',
       'logPercent_NC_3mnth']]
log_NC_3mnth_df = log_NC_3mnth_df.dropna()
log_NC_3mnth_df

# save to csv
log_NC_3mnth_df.to_csv("log_NC_3mnth.csv")
```

```python
# create dfs of each target column with null values
logPercent_ND_3mnth = target_df[['country','PC 1', 'PC 2', 'PC 3', 'Class',
       'logPercent_ND_3mnth']]
logPercent_ND_3mnth = logPercent_ND_3mnth.dropna()
logPercent_ND_3mnth

# save to csv
logPercent_ND_3mnth.to_csv("log_ND_3mnth.csv")
```

```python
# create dfs of each target column with null values
logPercent_TD_3mnth = target_df[['country','PC 1', 'PC 2', 'PC 3', 'Class',
       'logPercent_TD_3mnth']]
logPercent_TD_3mnth = logPercent_TD_3mnth.dropna()
logPercent_TD_3mnth

# save to csv
logPercent_TD_3mnth.to_csv("log_TD_3mnth.csv")
```
