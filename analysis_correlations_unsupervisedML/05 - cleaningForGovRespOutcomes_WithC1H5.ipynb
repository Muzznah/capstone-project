# Modification of dataset to include all feature & target variables for analysis

```python
# import dependencies
import pandas as pd
from datetime import timedelta as dt
```

```python
# load files
govRespOutcome = pd.read_csv("govRespOutcome_noC1H5.csv")
govReg = pd.read_csv("OxCGRT_latest(Aug31,2020).csv")

# preview dataframes (df)
display(govRespOutcome)
display(govReg)
print(list(govReg.columns))
```

```python
# rename column
govReg.rename(columns={"CountryName":"location"}, inplace=True)
govReg
```

```python
# fill null with zero
govReg = govReg.fillna(0)

# drop unneeded columns
govReg.drop(columns=['CountryCode', 'RegionName', 'RegionCode', 'Date', 'M1_Wildcard', 'ConfirmedCases', 'ConfirmedDeaths', 'StringencyIndex', 'StringencyIndexForDisplay', 'StringencyLegacyIndex', 'StringencyLegacyIndexForDisplay', 'GovernmentResponseIndex', 'GovernmentResponseIndexForDisplay', 'ContainmentHealthIndex', 'ContainmentHealthIndexForDisplay', 'EconomicSupportIndex', 'EconomicSupportIndexForDisplay'], inplace=True)

# check df
display(govReg)
govReg.dtypes
```

```python
# make df of non-monetary government measures
govReg_notMoney = govReg[['location','C1_School closing', 'C1_Flag', 'C2_Workplace closing', 'C2_Flag', 'C3_Cancel public events', 'C3_Flag', 'C4_Restrictions on gatherings', 'C4_Flag', 'C5_Close public transport', 'C5_Flag', 'C6_Stay at home requirements', 'C6_Flag', 'C7_Restrictions on internal movement', 'C7_Flag', 'C8_International travel controls', 'E1_Income support', 'E1_Flag', 'E2_Debt/contract relief', 'H1_Public information campaigns', 'H1_Flag', 'H2_Testing policy', 'H3_Contact tracing','H5_Investment in vaccines']]
display(govReg_notMoney)                         
GRNotMoney_grp = govReg_notMoney.groupby('location').max()
GRNotMoney_grp = GRNotMoney_grp.reset_index()
GRNotMoney_grp
```

```python
# make df of monetary gov measures
govReg_isMoney = govReg[['location','E3_Fiscal measures','E4_International support','H4_Emergency investment in healthcare']]
display(govReg_isMoney)                         
GRIsMoney_grp = govReg_isMoney.groupby('location').sum()
GRIsMoney_grp = GRIsMoney_grp.reset_index()
GRIsMoney_grp
```

```python
# merge non-monetary and monetary government measures dfs
govRespOutcomeNotMoney = govRespOutcome.merge(GRNotMoney_grp, on="location", how='left')
display(govRespOutcomeNotMoney)
govRespOutcomeIsMoney = govRespOutcomeNotMoney.merge(GRIsMoney_grp, on="location", how="left")
govRespOutcomeIsMoney
```

```python
# check df (column names, data types (dtypes))
pd.set_option("display.max_rows", 63)
print(govRespOutcomeIsMoney.dtypes)
govRespOutcomeIsMoney.columns
```

```python
# convert the following columns to datetime:
# minDate
# maxDate
# newCases_startDate
# MaxNewCases_date
# MaxNewDeaths_date
# MaxNewDeaths_date
# 3mth_date

govRespOutcomeIsMoney['minDate'] = pd.to_datetime(govRespOutcomeIsMoney['minDate'])
govRespOutcomeIsMoney['maxDate'] = pd.to_datetime(govRespOutcomeIsMoney['maxDate'])
govRespOutcomeIsMoney['newCases_startDate'] = pd.to_datetime(govRespOutcomeIsMoney['newCases_startDate'])
govRespOutcomeIsMoney['MaxNewCases_date'] = pd.to_datetime(govRespOutcomeIsMoney['MaxNewCases_date'])
govRespOutcomeIsMoney['MaxNewDeaths_date'] = pd.to_datetime(govRespOutcomeIsMoney['MaxNewDeaths_date'])
govRespOutcomeIsMoney['3mth_date'] = pd.to_datetime(govRespOutcomeIsMoney['3mth_date'])

govRespOutcomeIsMoney.dtypes
```

```python
# convert the following columns to integer:
# ramp_up_duration
# CaseVsResp_Interval
# StartToMaxNewCases_Interval
# MaxNewCasesToMaxNewDeaths_Interval
# FirstNewCaseToMaxNewDeaths

govRespOutcomeIsMoney['ramp_up_duration'] = pd.to_timedelta(govRespOutcomeIsMoney['ramp_up_duration']).dt.days
govRespOutcomeIsMoney['CaseVsResp_Interval'] = pd.to_timedelta(govRespOutcomeIsMoney['CaseVsResp_Interval']).dt.days
govRespOutcomeIsMoney['StartToMaxNewCases_Interval'] = pd.to_timedelta(govRespOutcomeIsMoney['StartToMaxNewCases_Interval']).dt.days
govRespOutcomeIsMoney['MaxNewCasesToMaxNewDeaths_Interval'] = pd.to_timedelta(govRespOutcomeIsMoney['MaxNewCasesToMaxNewDeaths_Interval']).dt.days
govRespOutcomeIsMoney['FirstNewCaseToMaxNewDeaths'] = pd.to_timedelta(govRespOutcomeIsMoney['FirstNewCaseToMaxNewDeaths']).dt.days

pd.set_option("display.max_columns", 63)
display(govRespOutcomeIsMoney)
```

```python
# view columns to prepare for column reorganization
govRespOutcomeIsMoney.columns.to_list()
```

```python
# reorganize columns
columns = ['primary_key_reaction','location','iso_code','population','population_density','median_age','gdp_per_capita',
           'minDate','maxDate','newCases_startDate','MaxNewCases_date','MaxNewDeaths_date','3mth_date',
           'ramp_up_duration','CaseVsResp_Interval','StartToMaxNewCases_Interval','FirstNewCaseToMaxNewDeaths','MaxNewCasesToMaxNewDeaths_Interval',
           'maxValue','AvgMaxGovResp_StartApr1','changeGovResp','StringencyIndex_updated','ContainmentHealthIndex_updated','EconomicSupportIndex_updated',
           'MaxNewCases_value','MaxNewDeaths_value','newCases_3mnth','totCases_3mnth','newDeaths_3mnth','totDeaths_3mnth',
           'residential','avgOutdoorMobil','retail_and_recreation','parks','grocery_and_pharmacy','transit_stations','workplaces',
           'C1_School closing', 'C1_Flag', 'C2_Workplace closing', 'C2_Flag',
       'C3_Cancel public events', 'C3_Flag', 'C4_Restrictions on gatherings',
       'C4_Flag', 'C5_Close public transport', 'C5_Flag',
       'C6_Stay at home requirements', 'C6_Flag',
       'C7_Restrictions on internal movement', 'C7_Flag',
       'C8_International travel controls', 'E1_Income support', 'E1_Flag',
       'E2_Debt/contract relief', 'H1_Public information campaigns', 'H1_Flag',
       'H2_Testing policy', 'H3_Contact tracing', 'H5_Investment in vaccines',
       'E3_Fiscal measures', 'E4_International support',
       'H4_Emergency investment in healthcare']

print(columns)
```
