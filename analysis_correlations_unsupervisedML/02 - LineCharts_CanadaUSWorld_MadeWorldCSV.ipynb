# This notebook created "world.csv"

```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
# load files
gov_reg_data = "government_regulation_FINAL.csv"
covid_stats_data = "covid_stats_FINAL.csv"
merged_covid_data = "merged_covid_FINAL.csv"
T_canUS_data = "T_canUS.csv"

gov_reg = pd.read_csv(gov_reg_data)
covid_stats = pd.read_csv(covid_stats_data)
merged_covid = pd.read_csv(merged_covid_data)
T_canUS_df = pd.read_csv(T_canUS_data)
```

# Create World dataframe

```python
# df of columns I want totals for
case_df = merged_covid[['date_updated','new_cases',
       'new_deaths', 'new_tests', 'total_cases_updated',
       'total_deaths_updated', 'total_tests_updated']]

# group by date to get totals
grpCase_df = case_df.groupby(by='date_updated').sum()

display(case_df.describe())
display(grpCase_df.describe())
display(grpCase_df)
```

```python
# df of columns I want averages for
index_df = merged_covid[['date_updated','StringencyIndex_updated', 'StringencyLegacyIndex_updated',
       'GovernmentResponseIndex_updated', 'ContainmentHealthIndex_updated',
       'EconomicSupportIndex_updated']]
display(index_df)

# group by date to get totals
grpIndex_df = index_df.groupby(by='date_updated').mean()

display(index_df.describe())
display(grpIndex_df.describe())
display(grpIndex_df)
```

```python
# join grouped grpCase and grpIndex dfs to create world_df
world_df = grpCase_df.join(grpIndex_df, how="inner")
world_df

# save world_df to csv
world_df.to_csv("world.csv")
```

```python
# preview T_canUS_df
display(T_canUS_df)

# make index 'date_updated'
T_canUS_df.set_index('date_updated', inplace=True)

# drop unnamed index
T_canUS_df.drop(columns="Unnamed: 0", inplace=True)
```

```python
# join world_df & T_canUS_df
canUSWorld_df = T_canUS_df.join(world_df, how="inner")
canUSWorld_df
```

```python
# rename world columns
canUSWorld_df = canUSWorld_df.rename(columns={"new_cases":"new_cases_W",
                             "new_deaths":"new_deaths_W",
                             "new_tests":"new_tests_W",
                             "total_cases_updated":"total_cases_updated_W",
                             "total_deaths_updated":"total_deaths_updated_W",
                             "total_tests_updated":"total_tests_updated_W",
                             "StringencyIndex_updated":"StringencyIndex_updated_W",
                             "StringencyLegacyIndex_updated":"StringencyLegacyIndex_updated_W",
                             "GovernmentResponseIndex_updated":"GovernmentResponseIndex_updated_W",
                             "ContainmentHealthIndex_updated":"ContainmentHealthIndex_updated_W",
                             "EconomicSupportIndex_updated":"EconomicSupportIndex_updated_W",
                             })

display(canUSWorld_df)
display(canUSWorld_df.columns)
```

# Scale world

```python
# scale: new cases, new deaths, new tests, total cases updated, total deaths updated, total tests updated
# for world

scaler = MinMaxScaler()

# fit scaler to df
colsNeedScale_df = canUSWorld_df[['new_cases_W', 'new_deaths_W', 'new_tests_W',
       'total_cases_updated_W', 'total_deaths_updated_W',
       'total_tests_updated_W']]

scaler.fit(colsNeedScale_df)

display(colsNeedScale_df.describe())
print(scaler.data_max_) # highest max value is total_tests_updated_W

# create series for each column that needs scaling
new_cases_W = canUSWorld_df['new_cases_W']
new_deaths_W = canUSWorld_df['new_deaths_W']
new_tests_W = canUSWorld_df['new_tests_W']
total_cases_updated_W = canUSWorld_df['total_cases_updated_W']
total_deaths_updated_W = canUSWorld_df['total_deaths_updated_W']
total_tests_updated_W = canUSWorld_df['total_tests_updated_W']
```

```python
# prep total_tests_updated_W to be fit by scaler
total_tests_updated_W = np.asarray(total_tests_updated_W)

total_tests_updated_W = total_tests_updated_W.reshape(-1,1)

# re-scale based on column with max value (total_tests_updated_W)
scaler.fit(total_tests_updated_W)
print(scaler.data_max_)

# transform total_tests_updated_W
T_total_tests_updated_W = scaler.transform(total_tests_updated_W)
print(T_total_tests_updated_W)
```

```python
# function that: applies asarray, reshape, transform to other columns
def scale(col):
    col = np.asarray(col)
    col = col.reshape(-1,1)
    col = scaler.transform(col)
    return col

# apply function
T_new_cases_W = scale(new_cases_W)
T_new_deaths_W = scale(new_deaths_W)
T_new_tests_W = scale(new_tests_W)
T_total_cases_updated_W = scale(total_cases_updated_W)
T_total_deaths_updated_W = scale(total_deaths_updated_W)

# drop old columns in T_canUS_df
T_canUSWorld_df = canUSWorld_df.copy()
T_canUSWorld_df.drop(columns=['new_cases_W', 'new_deaths_W', 'new_tests_W',
       'total_cases_updated_W', 'total_deaths_updated_W',
       'total_tests_updated_W'], inplace=True)

# attach new columns
T_canUSWorld_df['T_new_cases_W'] = T_new_cases_W
T_canUSWorld_df['T_new_deaths_W'] = T_new_deaths_W
T_canUSWorld_df['T_new_tests_W'] = T_new_tests_W
T_canUSWorld_df['T_total_cases_updated_W'] = T_total_cases_updated_W
T_canUSWorld_df['T_total_deaths_updated_W'] = T_total_deaths_updated_W
T_canUSWorld_df['T_total_tests_updated_W'] = T_total_tests_updated_W

# reset index
T_canUSWorld_df = T_canUSWorld_df.reset_index()

display(T_canUSWorld_df)
list(T_canUSWorld_df.columns)

pd.set_option('display.max_columns', None)
display(T_canUSWorld_df.describe())
```
