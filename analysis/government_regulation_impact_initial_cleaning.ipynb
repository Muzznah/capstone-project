```python
# In all "new" columns, replace NaN with 0 
covid_stats_df['new_cases'].fillna(0, inplace=True)
covid_stats_df['new_deaths'].fillna(0, inplace=True)
covid_stats_df['new_tests'].fillna(0, inplace=True)
covid_stats_df
```

```python
# In all "total_cases" columns, replace NaN with proceeding value 
locations = list(covid_stats_df['location'].unique())

DFlist = []

for location in locations:
    newDF = covid_stats_df.loc[covid_stats_df['location'] == location]
    
    total_cases = newDF['total_cases']
    total_deaths = newDF['total_deaths']
    total_tests = newDF['total_tests']
    
    filled_total_cases = total_cases.fillna(method="ffill")
    filled_total_deaths = total_deaths.fillna(method="ffill")
    filled_total_test = total_tests.fillna(method="ffill")
    
    newDF['total_cases_updated'] = filled_total_cases
    newDF['total_deaths_updated'] = filled_total_deaths
    newDF['total_tests_updated'] = filled_total_test
    
    newDF.drop(columns=['total_cases'], inplace=True)
    newDF.drop(columns=['total_deaths'], inplace=True)
    newDF.drop(columns=['total_tests'], inplace=True)
    
    DFlist.append(newDF)
    covid_stats_CLEAN = pd.concat(DFlist)
    
display(covid_stats_CLEAN)
```

```python
# It was discovered that for some countries there are is not any data on testing, will replace NaN with 0
covid_stats_CLEAN['total_tests_updated'].fillna(0, inplace=True)
covid_stats_CLEAN
```

### Merge and Export

```python
# Merge both DataFrames, drop redundant columns
merged_covid_df = pd.merge(covid_stats_CLEAN,government_regulation_CLEAN, how="left", left_on=['date_updated','location'], right_on=['Date_updated','CountryName'])
merged_covid_df.drop(['Date_updated','CountryName','CountryCode'],axis=1,inplace=True)
merged_covid_df
```
